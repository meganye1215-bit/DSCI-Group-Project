## Q: Can player experience level and age be used to predict the played hours of a player?

```r
players <- read_csv("data/players.csv")
head(players)
```

```r
players_clean <- players |> 
    clean_names()
head(players_clean)
tail(players_clean)
```

```r
nrow(players_clean)
```

```r
players_clean |> filter(is.na(age))
```

Since there are 2 missing values against total of 196 rows, we can filter these out

```r
players_filled <- players_clean |> filter(!is.na(age))
head(players_filled)
tail(players_filled)
```

Filter the observations where the played hours = 0

```r
players_tidy <- players_filled |> filter(played_hours != 0)
head(players_tidy)
tail(players_tidy)
```

Select the appropriate features, player experience level, age, and played hours.

```r
players_ds <- players_tidy |>
    select(experience, played_hours, age)
head(players_ds)
tail(players_ds)
```

Convert Experience to a categorical variable

```r
players_ds <- players_ds |>
    # mutate(experience=as_factor(experience))
    mutate(experience = factor(experience,
                              levels = c("Beginner", "Amateur", "Regular", "Veteran", "Pro"))) |>
    mutate(experience = as.numeric(experience))
head(players_ds)
tail(players_ds)
```

```r
options(repr.plot.width = 10, repr.plot.height = 8)
age_plot <- players_ds |> 
    ggplot(aes(x=age)) +
    geom_histogram(binwidth=3, fill = "skyblue", color = "black") +
    labs(
        x="Age (years)",
        y="Number of Players",
        title="Distribution of Player Ages"
    ) +
    theme(text = element_text(size=20))
age_plot
```

```r
options(repr.plot.width = 6, repr.plot.height = 8)
players_experience_plot <- players_ds |> 
    group_by(experience) |>
    summarize(count = n()) |>
    ggplot(aes(x=experience, y=count)) +
    geom_bar(stat="identity", fill = "skyblue", color = "black") +
    labs(
        x="Experience",
        y="Number of Players",
        title="Distribution of Experience"
    ) +
    theme(text = element_text(size=20))
players_experience_plot
```

```r
options(repr.plot.width = 12, repr.plot.height = 8)
played_hours_plot <- players_ds |> 
    ggplot(aes(x=played_hours)) +
    geom_histogram(binwidth=3, fill = "skyblue", color = "black") +
    labs(
        x="Played Hours (hr)",
        y="Number of Players",
        title="Distribution of Played Hours"
    ) +
    theme(text = element_text(size=20))
played_hours_plot
```

```r
options(repr.plot.width = 12, repr.plot.height = 8)
played_hours_plot <- players_ds |> 
    ggplot(aes(x=age, y=played_hours)) +
    geom_point(alpha=0.6) +
    labs(
        x="Age",
        y="Played Hours",
        title="Distribution of Age vs Played Hours"
    ) +
    theme(text = element_text(size=20))
played_hours_plot
```

We can see there are a few values which are abnormally high in the above graphic visual, played_hours > 150. We may choose to remove them.

### Prepare a knn regression model

```r
set.seed(2000)
players_split <- initial_split(players_ds, prop = 0.75, strata = played_hours)
players_training <- training(players_split)
players_testing <- testing(players_split)
```

```r
set.seed(1234)
players_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |> 
      set_engine("kknn") |>
      set_mode("regression") 

players_recipe <- recipe(played_hours ~ age + experience, data = players_training) |>
      step_scale(all_predictors()) |>
      step_center(all_predictors())
```

```r
set.seed(1234)
players_vfold <- vfold_cv(
    players_training,
    v=5,
    strata=played_hours
)
players_workflow <- workflow() |>
    add_model(players_spec) |>
    add_recipe(players_recipe)
players_workflow
```

```r
set.seed(2019)
gridvals <- tibble(neighbors = seq(1, 20, by=1))

players_results <- players_workflow |>
    tune_grid(
        resamples = players_vfold,
        grid = gridvals
    ) |>
    collect_metrics()
players_results
```

```r
set.seed(2020)
players_min <- players_results |>
    filter(.metric == "rmse") |>
    slice_min(mean, n=1) 
players_min
```

```r
set.seed(1234)
k_min <- players_min |>
            pull(neighbors)

players_best_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = k_min) |>
                            set_engine("kknn") |>
                            set_mode("regression")

players_best_fit <- workflow() |>
                        add_recipe(players_recipe) |>
                        add_model(players_best_spec) |>
                        fit(data = players_training)

players_shared_test_results <- players_best_fit |> 
                       predict(players_testing) |>
                       bind_cols(players_testing)

players_summary <- players_shared_test_results |>
                       metrics(truth = played_hours, estimate = .pred) 
players_shared_test_results
players_summary
```

### Remove the outliers visually seen in the played_hours distribution

```r
options(repr.plot.width = 12, repr.plot.height = 8)
played_hours_plot <- players_ds |> 
    ggplot(aes(x=played_hours)) +
    geom_histogram(binwidth=3, fill = "skyblue", color = "black") +
    labs(
        x="Played Hours (hr)",
        y="Number of Players",
        title="Distribution of Played Hours"
    ) +
    theme(text = element_text(size=20))
played_hours_plot
```

```r
players_no_outliers <- players_ds |>
    filter(played_hours < 150)

options(repr.plot.width = 12, repr.plot.height = 8)
played_hours_no_outliers_plot <- players_no_outliers |> 
    ggplot(aes(x=played_hours)) +
    geom_histogram(binwidth=3, fill = "skyblue", color = "black") +
    labs(
        x="Played Hours (hr)",
        y="Number of Players",
        title="Distribution of Played Hours"
    ) +
    theme(text = element_text(size=20))
played_hours_no_outliers_plot
```

```r
set.seed(2000)
players_no_outliers_split <- initial_split(players_no_outliers, prop = 0.75, strata = played_hours)
players_no_outliers_training <- training(players_no_outliers_split)
players_no_outliers_testing <- testing(players_no_outliers_split)
```

```r
set.seed(1234)
players_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |> 
      set_engine("kknn") |>
      set_mode("regression") 

players_recipe <- recipe(played_hours ~ age + experience, data = players_no_outliers_training) |>
      step_scale(all_predictors()) |>
      step_center(all_predictors())
```

```r
set.seed(1234)
players_vfold <- vfold_cv(
    players_no_outliers_training,
    v=5,
    strata=played_hours
)
players_workflow <- workflow() |>
    add_model(players_spec) |>
    add_recipe(players_recipe)
players_workflow
```

```r
set.seed(2019)
gridvals <- tibble(neighbors = seq(1, 10, by=1))

players_results <- players_workflow |>
    tune_grid(
        resamples = players_vfold,
        grid = gridvals
    ) |>
    collect_metrics()
players_results
```

```r
set.seed(2020)
players_min <- players_results |>
    filter(.metric == "rmse") |>
    slice_min(mean, n=1) 
players_min
```

```r
set.seed(1234)
k_min <- players_min |>
            pull(neighbors)

players_best_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = k_min) |>
                            set_engine("kknn") |>
                            set_mode("regression")

players_best_fit <- workflow() |>
                        add_recipe(players_recipe) |>
                        add_model(players_best_spec) |>
                        fit(data = players_no_outliers_training)

players_shared_test_results <- players_best_fit |> 
                       predict(players_no_outliers_testing) |>
                       bind_cols(players_no_outliers_testing)

players_summary <- players_shared_test_results |>
                       metrics(truth = played_hours, estimate = .pred) 
players_shared_test_results
players_summary
```
